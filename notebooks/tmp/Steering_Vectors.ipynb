```python
import torch
import tqdm

def get_sequence_log_perplexity(model, input_ids):
    max_length = model.config.n_positions
    stride = 512

    nlls = []
    for i in range(0, input_ids.size(1), stride):
        begin_loc = max(i + stride - max_length, 0)
        end_loc = min(i + stride, input_ids.size(1))
        trg_len = end_loc - i  # may be different from stride on last loop
        input_ids = input_ids[:, begin_loc:end_loc].to(model.device)
        target_ids = input_ids.clone()
        target_ids[:, :-trg_len] = -100

        outputs = model(input_ids, labels=target_ids)
        neg_log_likelihood = outputs[0] * trg_len

        nlls.append(neg_log_likelihood)

    ppl = torch.stack(nlls).sum() / end_loc
    return ppl
```

```python
def get_steering_vector(model, text, n_steps=100, lr=1e-4):
    input_ids = tokenizer.encode(text, return_tensors='pt')
    optimizer = torch.optim.Adam(model.transformer.steering.parameters(), lr=lr)
    w_init = torch.randn(768) / np.sqrt(768)
    model.transformer.steering.load_state_dict({"w": w_init})
    for i in tqdm.tqdm(range(n_steps)):
        loss = -get_sequence_log_perplexity(model, input_ids)
        loss.backward()
        optimizer.step()
        if i % 10 == 0:
            print("grad L1 norm: ", torch.mean(torch.abs(model.transformer.steering.w.grad)))
            print("difference between init and w:", torch.mean(torch.abs(w_init - model.transformer.steering.w.data)))
            print(loss.item())
    return model.transformer.steering.w.data.numpy()
```
